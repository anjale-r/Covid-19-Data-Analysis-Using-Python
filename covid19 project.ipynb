```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("D:/New folder/Python/Coursera/covid/covid19_Confirmed_dataset.csv")
data
```

```python
data.shape
```

Deleting useless columns

```python
data.drop(['Lat','Long'],axis=1,inplace=True)
```

```python
data.head()
```

Aggregating the rows by the country

```python
data_aggregated = data.groupby("Country/Region").sum()
```

```python
data_aggregated.head()
```

```python
data_aggregated.shape
```

Visualizing data related to a country for example China

```python
data_aggregated.loc['China'].plot()
data_aggregated.loc['Italy'].plot()
data_aggregated.loc['Spain'].plot()
plt.legend()
```

Calculating a good measure


we need to find a good measure reperestend as a number, describing the spread of the virus in a country.

```python
data_aggregated.loc['China'].plot()
```

ploting first derivative of the curve

```python
data_aggregated.loc['China'].diff().plot()
```

find maximum infection rate

```python
data_aggregated.loc['China'].diff().max()
```

```python
data_aggregated.loc['Italy'].diff().max()
```

```python
data_aggregated.loc['Spain'].diff().max()
```

max infection rate for all countries

```python
countries = list(data_aggregated.index)
max_infection_rates = []
for country in countries :
    max_infection_rates.append(data_aggregated.loc[country].diff().max())
data_aggregated['max infection rate'] = max_infection_rates
```

```python
data_aggregated.head()
```

```python
covid_data_new = pd.DataFrame(data_aggregated['max infection rate'])
```

```python
covid_data_new
```
